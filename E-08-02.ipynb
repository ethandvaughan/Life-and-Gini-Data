```python
def lifeFile(countryName):
    life = open("life.txt", "r")
    line = life.readline()
    countries = {}
    while line != "":
        lineList = line.strip()
        lineList = line.split()
        while lineList[2][0] != ":":
            lineList[1] = lineList[1] + " " + lineList[2]
            lineList.remove(lineList[2])
        lineList[2] = lineList[2][1:]
        countries[lineList[1]] = lineList[2]
        line = life.readline()
    try:
        print("  Life expectancy is", countries[countryName], "years at birth")
    except:
        print("  No Life Expectancy Data")
        
        
def giniFile(countryName):
    gini = open("gini.txt", "r")
    line = gini.readline()
    countries = {}
    while line != "":
        lineList = line.strip()
        lineList = line.split()
        while lineList[2][0] != ":":
            lineList[1] = lineList[1] + " " + lineList[2]
            lineList.remove(lineList[2])
        lineList[2] = lineList[2][1:]
        countries[lineList[1]] = lineList[2]
        line = gini.readline()
    try:
        print("  Gini value is", countries[countryName])
    except:
        print("  No Gini Data")
        
        
def suggest(countryName):
    gini = open("gini.txt", "r")
    line = gini.readline()
    countries = set()
    while line != "":
        lineList = line.strip()
        lineList = line.split()
        while lineList[2][0] != ":":
            lineList[1] = lineList[1] + " " + lineList[2]
            lineList.remove(lineList[2])
        countries.add(lineList[1])
        line = gini.readline()
        
    life = open("life.txt", "r")
    line = life.readline()
    while line != "":
        lineList = line.strip()
        lineList = line.split()
        while lineList[2][0] != ":":
            lineList[1] = lineList[1] + " " + lineList[2]
            lineList.remove(lineList[2])
        countries.add(lineList[1])
        line = life.readline()
        
    if countryName not in countries:
        print("  Maybr you meant:")
        countries = list(countries)
        countries = sorted(countries)
        for country in countries:
            if country[0].upper() == countryName[0].upper():
                print("  ", country)
```

```python
country = "a"
while country != "q":
    country = input("Please enter a country name ('q' to quit): ")
    if country != "q":
        lifeFile(country)
        giniFile(country)
        suggest(country)
```
